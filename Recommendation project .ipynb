حالا میخواهیم در ستون تایتل از دیتاست فیلم ها تاریخ ساخت را حذف کنیم. این کار با استفاده از رجکسی که شامل یه  پرانتز 4 رقم و یک پرانتز دیگر می شود میتوان تارریخ را جدا کرد و به یک ستون جدید اضافه کرد.

```python
movies_df['year'] = movies_df.title.str.extract('(\(\d\d\d\d\))',expand=False)
movies_df['year'] = movies_df.year.str.extract('(\d\d\d\d)',expand=False)
movies_df['title'] = movies_df.title.str.replace('(\(\d\d\d\d\))', '', regex=True)
#فضای خالی ابتدا و انتهای عنوان فیلم حذف می شود
movies_df['title'] = movies_df['title'].apply(lambda x: x.strip())
#پنج ردیف اول را برای چک کردن دستور صحیحی که دادیم جک میکنیم

movies_df.head()
```

حالا میخواهیم ژانرهای مختلف را از هم جدا کنیم

```python
#چون ژانرها با یک اسلش ساده از هم جدا شده اند میتوانیم از دستور اسپلیت به راحتی استفاده کنیم
movies_df['genres'] = movies_df.genres.str.split('|')
movies_df.head()
```

ر ادامه میخواهیم تمامی ژانر ها را به دیتاست به عنوان یک ستون اضافه کنیم و در صورت قرار گرفتن آن فیلم در ژانر عدد 1 و در صورت عدم قرار گرفتن عدد 0 بگذاریم

```python
#ابتدا یک کپی گرفته می شود
moviesWithGenres_df = movies_df.copy()

#حال دیتاست ردیف به ردیف خوانده شده و به ازای هر ژانر جدید یک ستون اضافه می شود و در صورت قرار گرفتن در آن ژانر عدد یک گذاشته می شود
for index, row in movies_df.iterrows():
    for genre in row['genres']:
        moviesWithGenres_df.at[index, genre] = 1
        
# برای سایر هم عدد صفر

moviesWithGenres_df = moviesWithGenres_df.fillna(0)

#نمایش 5 رکورد اول اطلاعات دیتافریم فیلم ها همراه با ژانرهای هر فیلم 
moviesWithGenres_df.head()
```

```python
#به نظرمی آید ستون زمان فیلم کاربردی هنیست پس آن را حذف میکنیم
ratings_df = ratings_df.drop(columns = 'timestamp')
ratings_df.head()
```

حالا اینپوت یک کاربر جدید که می خواهیم به او فیلم پیشنهاد دهیم را تعرف میکنیم. کاربر ما 5 فیلم دیده و به هر کدام امتیازی داده است:

```python
userInput = [
            {'title':'Breakfast Club, The', 'rating':5},
            {'title':'Toy Story', 'rating':3.5},
            {'title':'Jumanji', 'rating':2},
            {'title':"Pulp Fiction", 'rating':5},
            {'title':'Akira', 'rating':4.5}
         ] 
inputMovies = pd.DataFrame(userInput)
inputMovies
```

حالا میخواهیم آیدی فیلم هایی که کاربر دیده به دیتاست او اضافه شود

```python
#فیلتر دیتاست براساس فیلم هایی که کاربر دیده اسن
inputId = movies_df[movies_df['title'].isin(inputMovies['title'].tolist())]


#مرج ستون های حاوی اطلاعات به دیتاست کاربر
inputMovies = pd.merge(inputId, inputMovies)


#حذف ژانر و سال 
inputMovies = inputMovies.drop(columns = 'genres').drop(columns = 'year')

inputMovies
```

```python
#حالا ژانرها را هم اضافه میکنیم
userMovies = moviesWithGenres_df[moviesWithGenres_df['movieId'].isin(inputMovies['movieId'].tolist())]

userMovies
```

```python

#شاخص و یا ایندکس دیتافریم ایجاد شده ریست میگردد تا به ترتیب از عدد 0 شروع شود و یک واحد یک واحد اضافه گردد
userMovies = userMovies.reset_index(drop=True)


# از دیتافریم در دست بررسی ستون های شماره فیلم، عنوان فیلم، لیست ژانرها و سال تولید فیلم حذف میگردد
# دیتافریم ایجاد شده شامل ژانرهای هر فیلم میباشد (فیلم های دیده شده تویط کاربر)
userGenreTable = userMovies.drop(columns = 'movieId').drop(columns = 'title').drop(columns = 'genres').drop(columns = 'year')

userGenreTable
```

```python
# نمایش امتیاز های داده شده تویط کاربر به هر فبلم
inputMovies['rating']
# انجام ضرب ماتریسی در لیست ژانرهای هر فیلم و امتیازهای داده شده به هر فیلم توسط کاربر
userProfile = userGenreTable.transpose().dot(inputMovies['rating'])

#ماتریس حاصل پروفایل کاربر است
userProfile
```

```python
# در این مرحله ایندکس و یا شاخص هر فیلم را برابر با مقدار شماره فیلم میکند
genreTable = moviesWithGenres_df.set_index(moviesWithGenres_df['movieId'])

#حذف ستون های غیر ضروری

genreTable = genreTable.drop(columns = 'movieId').drop(columns = 'title').drop(columns = 'genres').drop(columns = 'year')

genreTable.head()
```

```python
#(genreTable*userProfile)
# ضرب امتیار کاربر به هر ژانر در ماتریس ژانرهای فیلم ها

#(genreTable*userProfile).sum(axis=1)
# جمع امتیار به دست آماده برای هر فیلم در تنامی ژانرهای آن فیلم

#userProfile.sum()
# جمع امتیار پروفایل کاربر 


# محاسبه امتیار هر فیلم براساس ژانرهای آن فیلم و تقسیم به جمع امتیارات پروفایل کاربر
recommendationTable_df = ((genreTable*userProfile).sum(axis=1))/(userProfile.sum())

# نمایش امتیازات هر فیلم (نمایش 5 فیلم ابتدایی)
recommendationTable_df.head()
```

```python
#سورت لیست فیلم ها به صورت نزولی
recommendationTable_df = recommendationTable_df.sort_values(ascending=False)
recommendationTable_df.head()
```
